```python
import yfinance as yf
import pandas as pd

nyse_asset = '^NYA'
min_year = 2007
max_year = 2022
```

```python
nyse_index = yf.Ticker(f'{nyse_asset}')
```

```python
# Histórico de 1 ano com intervalo diário
start = '2006-01-01'
end = '2022-12-31'

nyse_history = nyse_index.history(start=start, end=end).reset_index()
```

```python
(nyse_history
    .assign(Year=lambda x: x['Date'].dt.year)
    .groupby('Year')
    .mean()
    .reset_index()
    .filter(items=['Date', 'Year', 'Close'])
    #.to_csv('files/nyse_index_data_2006-2022.csv')
)
```

```python
nyse_history = pd.read_csv('files/nyse_index_data_2006-2022.csv')
nyse_history = nyse_history.query(f'Year >= {min_year} & Year <= {max_year}').drop('Unnamed: 0', axis=1)
nyse_history
```

```python
ipc_usa = pd.read_excel('files/ipc_usa_2007-2022.xlsx', skiprows=11, engine='openpyxl')
ipc_usa
```

```python
ipc_usa = (ipc_usa
    .filter(items=['Year', 'HALF1', 'HALF2'])
    .assign(Mean=lambda x: (x['HALF1'] + x['HALF2']) / 2)
)

ipc_usa
```

```python
gold_price = pd.read_csv('files/gold_price.csv')
gold_price = gold_price.query(f'Ano >= {min_year} & Ano <= {max_year}').drop('Unnamed: 0', axis=1)
gold_price
```

```python
(gold_price
    .merge(right=ipc_usa, how='left', left_on='Ano', right_on='Year')
    .filter(items=['Ano', 'Último', 'Mean'])
    .rename(columns={
        'Ano': 'ano',
        'Último': 'preco_ouro',
        'Mean': 'ipc'
    })
    .merge(right=nyse_history, how='left', left_on='ano', right_on='Year')
    .filter(items=['ano', 'preco_ouro', 'ipc', 'Close'])
    .rename(columns={
        'Close': 'indice_nyse'
    })
    #.to_csv('dataset/gold_dataset.csv')
)
```
